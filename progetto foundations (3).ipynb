## 7- Cluster the pickup time of the day into 30-minute intervals (e.g. from 02:00 to 02:30)

```python

intervals= list(pd.date_range(start = "00:00", end= "23:30", freq = "30min").strftime("%H:%M")) #end="24:00" makes some issues, we add 24:00 manually
intervals.append("00:00")
labels = [f"{intervals[i]}-{intervals[i + 1]}" for i in range(len(intervals) - 1)]

data['30_min_interval'] = pd.cut(
    #converts the pick up time into minutes to assign it to the minute interval corresponding to the label  
    pd.to_datetime(data['tpep_pickup_datetime'], format='%H:%M:%S').dt.hour * 60 + pd.to_datetime(data['tpep_pickup_datetime'], format='%H:%M:%S').dt.minute,
    bins=range(0, 1441, 30), #24*60 + 1 limits of interval in minutes 
    labels=labels, #label of interval
    right=False #interval closed on the left and open on the right
)

data[['tpep_pickup_datetime', '30_min_interval']]
```

## 13 - Build a new dataframe (called common) where, for each pickup location we keep all trips to the 5 most common destinations (i.e. each pickup location can have different common destinations).

```python
trips_count = data.groupby(['PULocationID', 'DOLocationID']).size().reset_index(name='trip_count')
trips_count = trips_count.sort_values(['PULocationID', 'trip_count'], ascending=[True, False])
common = trips_count.groupby('PULocationID').apply(lambda group: group.head(5)).reset_index(drop=True) #head(5) first 5 rows for each PULocationID
common
```

## 14 - On the common dataframe, for each payment type and each interval, determine the average fare amount

```python
# merge common with data to concat the values 
#we use again the df with fare_amount>=0
common = data_fare_amount.merge(
    common,  # Merge with top destinations
    on=['PULocationID', 'DOLocationID'],  # Match on PULocationID and DOLocationID
    how='inner'  # Retain only matching rows
)
common = common.drop_duplicates(subset=['PULocationID', 'DOLocationID'])

common = common.sort_values(['PULocationID', 'trip_count'], ascending=[True, False])

common
```

## 17 - Build chains of trips. Two trips are consecutive in a chain if (a) they have the same VendorID, (b) the pickup location of the second trip is also the dropoff location of the first trip, (c) the pickup time of the second trip is after the dropoff time of the first trip, and (d) the pickup time of the second trip is at most 2 minutes later than the dropoff time of the first trip.

```python

data_sorted = data.sort_values(by=["VendorID", "tpep_dropoff_datetime", "tpep_pickup_datetime"])

# We create columns to compare the current trip with the previous one
data_sorted["prev_DOLocationID"] = data_sorted["DOLocationID"].shift(1)
data_sorted["prev_VendorID"] = data_sorted["VendorID"].shift(1)
data_sorted["prev_tpep_dropoff_datetime"] = data_sorted["tpep_dropoff_datetime"].shift(1)
data_sorted["prev_PULocationID"] = data_sorted["PULocationID"].shift(1)

# time difference between pickup e dorpoff of 2 different trips
data_sorted["time_diff"] = (data_sorted["tpep_pickup_datetime"] - data_sorted["prev_tpep_dropoff_datetime"]).dt.total_seconds()

# Chain condition
is_chain = (
    (data_sorted["VendorID"] == data_sorted["prev_VendorID"]) &  # Same VendorID
    (data_sorted["PULocationID"] == data_sorted["prev_DOLocationID"]) &  # current Pickup  = previous Dropoff 
    (data_sorted["time_diff"].between(0, 120))  
) 

# Let's create a column that identifies the chains
data_sorted["chain_id"] = (is_chain == False).cumsum() #True start of a chain, False represents consecutive trips that are part of the same chain.

data_sorted[["VendorID", "PULocationID", "DOLocationID", "tpep_pickup_datetime", "tpep_dropoff_datetime", "chain_id"]]
```
