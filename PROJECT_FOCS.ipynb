5- Add a duration column storing how long each trip has taken (use tpep_pickup_datetime, tpep_dropoff_datetime)

```python
data["tpep_pickup_datetime"]=pd.to_datetime(data["tpep_pickup_datetime"], format='mixed')
data["tpep_dropoff_datetime"]=pd.to_datetime(data["tpep_dropoff_datetime"], format='mixed')
```

```python
data["trip_duration"] = data["tpep_dropoff_datetime"] - data["tpep_pickup_datetime"]
data
```

6- For each pickup location, determine how many trips have started there.

```python
data.groupby("PULocationID").size().reset_index(name='Numeber of trips') 
```

7- Cluster the pickup time of the day into 30-minute intervals (e.g. from 02:00 to 02:30)

```python

intervals= list(pd.date_range(start = "00:00", end= "23:30", freq = "30min").strftime("%H:%M"))
intervals.append("00:00")
labels = [f"{intervals[i]}-{intervals[i + 1]}" for i in range(len(intervals) - 1)]

data['30_min_interval'] = pd.cut(
    #convert minutes to put the total minutes in the corresponding minute range the corresponding label  
    pd.to_datetime(data['tpep_pickup_datetime'], format='%H:%M:%S').dt.hour * 60 + pd.to_datetime(data['tpep_pickup_datetime'], format='%H:%M:%S').dt.minute,
    bins=range(0, 1441, 30), #24*60 + 1 limits of interval in minutes 
    labels=labels,
    right=False #intervallo chiuso a sx e aperto a destra
)

# Visualizzazione del risultato
data[['tpep_pickup_datetime', '30_min_interval']]
```

8- For each interval, determine the average number of passengers and the average fare amount.

```python
data.groupby("30_min_interval").agg(
    Average_Passenger_Per_Trips=('passenger_count', 'mean'),   
    Average_Fare_Amount=('fare_amount', 'mean')             
).reset_index()
#agg to aggregate the function and rename the column easly
#we used this methon instead data.groupby("30_min_interval")[['passenger_count',''total_amount']].mean() to rename easly and write less rows of code
# no renames function
```

9- For each payment type and each interval, determine the average fare amount

```python
data.groupby(["payment_type","30_min_interval"])['fare_amount'].mean().reset_index(name="average fare amount")
```
